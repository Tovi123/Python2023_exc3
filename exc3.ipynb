```python
#api_key = 'Please insert your API KEY here'
api_key = "AIzaSyCUcFSaymLrKYtbPMpYXO16I93VkI2mjvg"
def destinations(origin,destination):
    url = f"https://maps.googleapis.com/maps/api/distancematrix/json?origins={origin}&destinations={destination}&units=kilometers&key={api_key}"
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error",response.status_code)
        else:
            try:
                response_data = response.json()
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")
        
    distance = (response_data)["rows"][0]["elements"][0]["distance"]["text"]
    target = (response_data)["destination_addresses"][0].split(",")[0]
    duration = (response_data)["rows"][0]["elements"][0]["duration"]["value"]
    hours,minutes = duration // 3600 , (duration % 3600) // 60
    duration= f"{hours} hours and {minutes} minutes"
    
    url="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (destination,api_key)
        
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error",response.status_code)
        else:
            try:
                response_data = response.json()
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")
    
    Longitude = response_data ["results"][0]["geometry"]["location"]["lng"]    
    latitude = response_data ["results"][0]["geometry"]["location"]["lat"]
    
    return target,distance,duration,Longitude,latitude
```

```python
my_dict = {}

# Create empty lists to hold the values for each column
target_list ,dist_list ,Dur_list ,long_list ,lat_list = [],[],[],[],[]

fname = open("dests.txt")

origin = "Tel Aviv"

# Read each line from the file and get the destinations
for line in fname:
    line = line.rstrip()
    target, distance,duration,lng, lat = destinations(origin, line)
    target_list.append(target)
    dist_list.append(distance)
    Dur_list.append(duration)
    long_list.append(lng)
    lat_list.append(lat)

# Add the lists to the dictionary
my_dict["Target"] = target_list
my_dict["Distance_km"] = dist_list
my_dict["Duration (hour+minutes)"] = Dur_list
my_dict["Longitude"] = long_list
my_dict["latitude"] = lat_list

# Convert the dictionary to a pandas dataframe
data = pd.DataFrame.from_dict(my_dict)
```
